```python
from thefuzz import process
```

```python
#from above cell output we can see that the spam email has some words in common like win cash prize,free gift card,discout limited offer,lotter free,account suspended, otp account urgent 
choices=[ 'win cash prize','free gift card','discout limited offer','lottery free','account suspended', 'otp account urgent']
data["matches"] = data["email_text"].apply(
    lambda x: process.extract(x, choices, limit=1)
)
```

```python
data["spam"] = data["matches"].apply(
    lambda x: int(x[0][1] > 85)
)
```
